```python
# Sub Plot

import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10,1)
y=2*x+5
y1=5*x+10

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)              #........ row,column,fig_no
plt.plot(x,y)
plt.title('graph1')

plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title('graph2')
plt.show()
```
