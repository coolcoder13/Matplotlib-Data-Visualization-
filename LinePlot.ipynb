```python
# Line Plot

import numpy as np
import matplotlib.pyplot as plt

f = lambda x : int(x)**2
f1 = lambda x : int(x)**3
a = [int(i) for i in range(10)]
b = [f(i) for i in a]
c = [f1(i) for i in a]

plt.plot(a,b,linewidth=2.0,color='c')
plt.stackplot(a,b,color='darkmagenta',alpha=0.7)
plt.plot(a,c,linewidth=1.0,color='r')
plt.stackplot(a,c,color='slategrey',alpha=0.5)
plt.show()
```
