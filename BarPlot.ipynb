```python
# Bar Plot

import numpy as np
import matplotlib.pyplot as plt

language = ['Python','Java','Kortlin','Php','C#','C']
myscores = [99,70,95,70,50,60]

colors=['#F44336','#E91E63','#9C27B0','#FFC107','#FF9800','#FF5722']

ax = plt.bar(x=language,height=myscores,color=colors,hatch='*')
for i,val in enumerate(myscores):
    plt.text(i-.15,val+1,str(val))
```
