```python
# Violin Plot
# gives the probability density function

import numpy as np
import matplotlib.pyplot as plt

total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
boolean = [0,1,0,1,1,1,0,1,1,0]

data=list([total,order,discount])
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.violinplot(data, showmeans=True, showmedians=True)
plt.grid(True)
plt.title('Violin Plot Demo')

plt.subplot(1,2,2)
plt.violinplot(boolean, showmeans=True, showmedians=True)
plt.grid(True)

plt.show()
```
