```python
# Donut

import numpy as np
import matplotlib.pyplot as plt

tweets = [5390,1894,955]
labels = ['positive','neautral','negetive']
colors = ['#2196F3','#FFEB3B','#F44336']
explode = [0,0,.2]

ax = plt.pie(x=tweets,
             labels=labels,
             colors=colors,
             autopct='%.1f%%',
             explode=explode,
             wedgeprops={'width':.3},
             pctdistance=.5,
             shadow=True
            )
plt.axis('equal')
plt.show()
```
