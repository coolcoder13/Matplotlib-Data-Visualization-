```python
# Customizing Line Plot

import numpy as np
import matplotlib.pyplot as plt

#defining dataset
# %matplotlib qt        ....... will make graph on a new window
x=np.arange(10)
y=2*x+5
y1=5*x+6

#creating figure
fig = plt.figure(figsize=(15,5))

#plotting the points
plt.plot(x,y,linewidth=2.0,linestyle=":",color='r',alpha=0.5,marker='o')      #alpha is for brightness of color
plt.plot(x,y1,linewidth=3.0,linestyle=':',color='b',alpha=0.8,marker='*')
plt.title('title of graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['line_1','line_2'],loc='best')                  #....loc best means it will put at best position in whole graph
plt.grid(True)
plt.show()
```
