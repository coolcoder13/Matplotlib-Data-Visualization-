```python
# Horizontal Bar Plot

import numpy as np
import matplotlib.pyplot as plt

data = {'apple':40,'mango':60,'banana':10,'orange':30}
x=list(data.keys())
y=list(data.values())

plt.figure(figsize=(20,5))
plt.title('fruits')
plt.ylabel('fruits_name')
plt.xlabel('rating')
plt.barh(x,y,color=['y','c','orange','maroon'])
plt.show()
```
