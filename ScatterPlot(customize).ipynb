```python
# Customized Scatter plot

import numpy as np
import matplotlib.pyplot as plt

a=np.arange(10,90,10)
b=[5,3,2,5,6,1,4,2]
x=[1,2,3,4,3,2,1,2]

plt.figure(figsize=(15,5))
plt.scatter(a,b,color='g',s=200,edgecolors='y',marker='o',alpha=0.5)
plt.scatter(a,x,c='maroon',s=400,edgecolors='b',marker='4',alpha=1)
plt.legend(['b','x'],loc='best')
plt.title('customized of scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.savefig('Custom Scatter.png')
plt.show()
```
