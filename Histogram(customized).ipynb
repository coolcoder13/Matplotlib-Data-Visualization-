```python
# Customized Histogram

import numpy as np
import matplotlib.pyplot as plt

# tells frequency...... here how many no.s are present b/w 0,19 20,39 and so on is counted
n=[10,12,16,14,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]

plt.figure(figsize=(15,8))
plt.hist(n , bins=[0,20,40,60,80,100],edgecolor='y',color='r')
plt.title('customized of histogram')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```
