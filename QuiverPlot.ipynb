```python
# Quiver Plot
# Quiver Plot are used to represent vectors and Stream plot are used to show +ve -ve values and both are mostly used in physics

import numpy as np
import matplotlib.pyplot as plt

x_pos=y_pos=[0,0]
x_direct=[1,0]
y_direct=[1,-1]

plt.figure(figsize=(10,5))
plt.quiver(x_pos,y_pos,x_direct,y_direct,scale=5)
plt.axis([-1.5 , 1.5 , -1.5 , 1.5])
plt.grid(True)
plt.show()
```
